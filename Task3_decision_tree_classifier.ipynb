> Classifying the right species class using the decision tree algorithm.



Importing the Datasets.

```
import pandas as pd
data = pd.read_csv('/content/Iris (2).csv')
data.head()
```

### Data Preparation

```
x = data.iloc[:,[0,1,2,3]].values
y= data[['Species']].values
```

```
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```
#encoding labels for the species
data["Species"]=label.fit_transform(data["Species"])
```

```
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

# Decision Tree Algorithm 

```
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy', max_depth=3)
dt.fit(xtrain,ytrain)

print('Decision Tree Classifer Created')
```

Prediction

```
ypred=dt.predict(xtest)
```

```
ypred
```

```
ytest
```

# Visualizing the Decsion Tree Plot.

```
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plot_tree(dt,filled=True)
plt.show()
```

## Metrics and Evaluation

```
# Accuracy of Classification
from sklearn.metrics import accuracy_score
print('Accuracy score:',accuracy_score(ytest,ypred))
```

```
#metrics 
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix
import seaborn as sns
```

```
#overall metrics and evaluation score
print(metrics.classification_report(ytest, ypred))
```
